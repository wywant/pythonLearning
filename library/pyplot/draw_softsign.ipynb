```python
import math
import matplotlib.pyplot as plt
import numpy as np
def softsign(x):
    return x / (np.abs(x) + 1)

x1 = np.arange(-1.0,1,0.01)
y1 = [softsign(x)  for x in x1]

x2 = np.arange(-3.0,3,0.01)
y2 = [softsign(x)  for x in x2]

x3 = np.arange(-10.0,10,0.01)
y3 = [softsign(x)  for x in x3]

x4 = np.arange(-100.0,100,0.01)
y4 = [softsign(x)  for x in x4]

plt.subplot(2, 2, 1)
plt.plot(x1, y1)

plt.subplot(2, 2, 2)
plt.plot(x2, y2)

plt.subplot(2, 2, 3)
plt.plot(x3, y3)

plt.subplot(2, 2, 4)
plt.plot(x4, y4)

plt.show()
```
