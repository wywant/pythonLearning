```python
# https://docs.python.org/3/library/inspect.html
import inspect
import tensorflow.keras.layers as layers
```

```python


#ls.append(layers.Layer)
#ls.append(layers.Flatten)
#ls.append(layers.Dense)
#ls.append(layers.Conv1D)
#ls.append(layers.Conv2D)
#ls.append(layers.Conv3D)
#ls.append(layers.ReLU)
#ls.append(layers.RNN)
#ls.append(layers.LSTM)

names = dir(layers)
classesInPackage = []
for name in names:
    attr = getattr(layers, name)
    if inspect.isclass(attr):
        #print(attr)
        classesInPackage.append(attr)
        #print(name)
layerSubClasses = []
for classInPackage in classesInPackage:
    if classInPackage.__base__ == layers.Layer:
        layerSubClasses.append(classInPackage)
        print(classInPackage)


cs = []
cnames = []
for alayer in layerSubClasses:
    cnames.append(alayer.__name__)
    cs.append(inspect.signature(alayer.__init__))
    
arguments = []
signatureArguments = []
for signature in cs:
    args = []
    for k in signature.parameters:
        if k == 'self':
            continue;
        if k == 'kwargs':
            continue;
        if k not in arguments:
            arguments.append(k)
        args.append(k)
    print(args)
    signatureArguments.append(args)

#print(cnames)             
lines = []
for k in arguments:
#    print(k)
    line = []
    line.append(k)
    for signatureArgument in signatureArguments:
        if k in signatureArgument:
            line.append('Y')
        else :
            line.append('')
    lines.append(line)
    print(','.join(line))

#    print(line)
with open('layers.csv', 'wt') as f:
    f.write(',')
    f.write(','.join(cnames))
    f.write('\n')
    for line in lines:
        f.write(','.join(line))
        f.write('\n')
    
        
#    print(k)
#    print(type(k))
#print(signature)
#print(signature.parameters)
#print(signature.args)
```
