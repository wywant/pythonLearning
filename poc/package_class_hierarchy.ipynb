```python
import inspect
import tensorflow.keras.layers as layers

super2sub = {}

names = dir(layers)
classesInPackage = []
for name in names:
    attr = getattr(layers, name)
    if inspect.isclass(attr):
        # print(name, attr)
        classesInPackage.append(attr)
        superClass = attr.__base__
      
        if not superClass:
            superClass = '_Blank'
      
        subs = super2sub.get(superClass)
        if not subs :
            subs = []
            super2sub[superClass] = subs
        subs.append(attr)    
        #print(attr)
#for s,v in super2sub.items():
#    print(s)

allSubs = []
classesOutOfPackage = []
for s,subs in super2sub.items():
    for sub in subs:
        if sub not in allSubs:
            allSubs.append(sub)

    if s in classesInPackage:
        continue
    else :
        classesOutOfPackage.append(s)
        #print(s)
        for sub in subs:
            #print('  ', sub)
            continue

def aaa(current, depth):
    print('  ' * depth, current)
    if current in super2sub.keys():
        subs = super2sub[current]
        for sub in subs:
            aaa(sub, depth + 1)

    
print('############')
for classOutOfPackage in classesOutOfPackage:
    aaa(classOutOfPackage, 0)
    

#aaa()
    
```
